```python
#importing Libraries
import pandas as pd
import numpy as np
from datetime import date
import openpyxl
```

<h4> 2.1 Cleaning the Fighter Data Set </h4>

```python
#import data sets for fighters
fightermetric_df = pd.read_csv(r"fightermetric.csv")
fighter_df = pd.read_csv(r"fighterdata.csv")
```

```python
fighter_df.head()
```

```python
print("Dataset Shape:", fighter_df.shape)
```

```python
print("Dataset Datatype\n", fighter_df.dtypes)
```

```python
print("Dataset Null Values\n",fighter_df.isna().sum())
```

```python
fighter_df['Age'] = pd.to_datetime(fighter_df['DOB'].str.strip())
fighter_df['Age'] = fighter_df['Age'].dt.year
```

```python
current_year = date.today().year
fighter_df['Age'] = current_year - fighter_df['Age']
```

```python
fighter_df['Age'] = fighter_df['Age'].fillna(0).astype(int)
fighter_df['Nickname'] = fighter_df['Nickname'].fillna('None')
fighter_df[['Height','Weight','Stance','Reach','DOB']] = fighter_df[['Height','Weight','Stance','Reach','DOB']].fillna('NA')
```

```python
fighter_df.head()
```

```python
fightermetric_df.head()
```

```python
fightermetric_df.shape
```

```python
#Merging the two datasets
mergeddata_df = pd.merge(fighter_df, fightermetric_df, left_on = 'Fighter Name', right_on = 'NAME', how = 'left')
```

```python
print("Dataset Null Values\n",mergeddata_df.isna().sum())
```

```python
mergeddata_df = mergeddata_df.drop(['NAME'],axis=1)
```

```python
mergeddata_df.head()
```

```python
#Null Values can be filled by scraping remaining data from scrapper program. Adding new csv.
scraped_data = pd.read_excel('scrapedfighterdata.xlsx')
```

```python
scraped_data.columns
```

```python
scraped_data = scraped_data.drop(['Unnamed: 0', 'DOB:', 'Height:', 'Reach:', 'STANCE:','Weight:'], axis = 1)
```

```python
scraped_data=scraped_data.rename(columns = {'SApM:':'SAPM', 'SLpM:':'SLPM','Str. Acc.:':'STRA','Str. Def:':'STRD',
                        'TD Avg.:':'TD','TD Acc.:':'TDA','TD Def.:':'TDD','Sub. Avg.:':'SUBA', 'Name':'Fighter Name'})
```

```python
scraped_data['STRA'] = scraped_data['STRA'].str.rstrip('%').astype('float') / 100.0
```

```python
scraped_data['STRD'] = scraped_data['STRD'].str.rstrip('%').astype('float') / 100.0
scraped_data['TDA'] = scraped_data['TDA'].str.rstrip('%').astype('float') / 100.0
```

```python
scraped_data['TDD'] = scraped_data['TDD'].str.rstrip('%').astype('float') / 100.0
```

```python
scraped_data.head()
```

```python
clean_df = pd.merge(mergeddata_df, scraped_data, left_on = 'Fighter Name', right_on = 'Fighter Name', how = 'left')
```

```python
clean_df.columns
```

```python
clean_df['SLPM_x'] = clean_df['SLPM_x'].fillna(clean_df['SLPM_y'])
clean_df['SAPM_x'] = clean_df['SAPM_x'].fillna(clean_df['SAPM_y'])
clean_df['STRA_x'] = clean_df['STRA_x'].fillna(clean_df['STRA_y'])
clean_df['STRD_x'] = clean_df['STRD_x'].fillna(clean_df['STRD_y'])
clean_df['TD_x'] = clean_df['TD_x'].fillna(clean_df['TD_y'])
clean_df['TDA_x'] = clean_df['TDA_x'].fillna(clean_df['TDA_y'])
clean_df['TDD_x'] = clean_df['TDD_x'].fillna(clean_df['TDD_y'])
clean_df['SUBA_x'] = clean_df['SUBA_x'].fillna(clean_df['SUBA_y'])
```

```python
clean_df = clean_df.drop(['SAPM_y', 'SLPM_y','STRA_y', 'STRD_y', 'SUBA_y', 'TDA_y', 'TD_y', 'TDD_y'], axis = 1)
```

```python
#writting Clean data to excel File
clean_df.to_excel("fighter_eda.xlsx")
```
